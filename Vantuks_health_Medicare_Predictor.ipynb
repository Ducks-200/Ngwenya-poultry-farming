```python
%%writefile app.py
import streamlit as st
import pandas as pd
import qrcode
from io import BytesIO
from sklearn.tree import DecisionTreeClassifier
import smtplib
from email.mime.text import MIMEText

# --- CONFIGURATION & BRANDING ---
st.set_page_config(page_title="Vantuka Health Medicare Predictor", layout="wide")

# Owner Details
OWNER_NAME = "MATUME SEBETOLA"
CONTACT_NO = "0699962430"
EMAIL = "matumesebetola@gmail.com"

# SMTP Configuration (For sending emails - IMPORTANT: Update these with your real details)
SMTP_SENDER_EMAIL = "your_email@gmail.com" # <--- CHANGE THIS TO YOUR GMAIL ADDRESS
SMTP_APP_PASSWORD = "your_16_char_app_password" # <--- CHANGE THIS TO YOUR GENERATED APP PASSWORD

# --- SIDEBAR NAVIGATION ---
st.sidebar.title("🏥 Vantuka Health")
menu = st.sidebar.selectbox("Navigate Menu", [
    "Home & About Us", "AI Medicare Predictor", "Guidelines & Training",
    "Hospitals & Pharmacy Finder", "Bookings & Mentorship",
    "Healthcare Tutoring & Jobs", "Patient & File Vault", "Contact & Inquiries"
])

# --- 1. HOME & ABOUT US ---
if menu == "Home & About Us":
    st.header("Welcome to Vantuka Health Medicare Predictor")
    st.subheader(f"Owner/Funder: {OWNER_NAME}")

    col1, col2 = st.columns(2)
    with col1:
        st.write("### Our Services")
        st.write("- Global Healthcare Data Prediction\n- Online Treatment Bookings\n- Professional Medical Training\n- AI Assistant Support")

    with col2:
        st.write("### About Us")
        st.info("Vantuka Health is a globalized AI-driven platform designed to bridge the gap between patients and healthcare providers.")

    # QR Code Generation for Account Access
    st.write("---")
    st.write("### Access Barcode")
    qr = qrcode.make(f"https://vantukahealth.com/login?owner={OWNER_NAME}")
    buf = BytesIO()
    qr.save(buf)
    st.image(buf, caption="Scan to access your account globally")

# --- 2. AI MEDICARE PREDICTOR ---
elif menu == "AI Medicare Predictor":
    st.header("🤖 AI Data & Medicine Predictor")

    tab1, tab2 = st.tabs(["Patient Diagnosis", "Doctor/Nurse Data Insights"])

    with tab1:
        st.write("### Symptoms Predictor")
        symptom = st.text_input("Enter Symptoms (e.g., Fever, Cough)")
        severity = st.select_slider("Severity Level", options=["Low", "Medium", "High"])

        if st.button("Predict Treatment"):
            st.success("Prediction: Based on current data, Paracetamol and hydration are recommended. (Consult a professional)")
            st.write("**Usage Instructions:** Take 1 tablet every 8 hours after meals.")

    with tab2:
        st.write("### Healthcare Professional Insights")
        st.write("Predictive analytics for medicine stock and patient recovery trends.")
        st.bar_chart({"Recovered": [20, 45, 80], "Pending": [10, 25, 5]})

# --- 3. GUIDELINES & TRAINING ---
elif menu == "Guidelines & Training":
    st.header("📚 Healthcare Manuals & Training")

    category = st.radio("Select Category", ["Medicine Usage", "Symptoms Guidelines", "Staff Training", "Caregivers/NGOs"])

    if category == "Staff Training":
        st.write("### Doctors & Nurses Training Portal")
        st.video("https://www.youtube.com/watch?v=dQw4w9WgXcQ") # Placeholder for tutorial
        if st.button("Download Study Guide (PDF)"):
            st.download_button("Download", data="Manual Content", file_name="Vantuka_Health_Guide.pdf")

    st.write("### Online Certification")
    st.write("Complete the module below to receive your Certificate of Completion.")
    if st.button("Start Certification Test"):
        st.write("Test Loading...")

# --- 4. HOSPITALS & PHARMACY FINDER ---
elif menu == "Hospitals & Pharmacy Finder":
    st.header("📍 Healthcare Network Locator")
    location = st.selectbox("Select Region", ["Gauteng", "Limpopo", "Western Cape", "Global/International"])

    # Mock Data
    hospitals = pd.DataFrame({
        'Name': ['Vantuka General', 'City Pharmacy', 'Sebetola Surgery'],
        'Address': ['14272 Ivory Park', 'Midrand Central', 'Pretoria East'],
        'Contact': ['011 123 4567', '011 999 8888', CONTACT_NO]
    })
    st.table(hospitals)

# --- 5. PATIENT & FILE VAULT ---
elif menu == "Patient & File Vault":
    st.header("📂 Secure File & Patient Saving")

    uploaded_file = st.file_uploader("Save Medical Files/Reports", type=['pdf', 'jpg', 'png', 'docx'])
    if uploaded_file:
        st.success(f"File '{uploaded_file.name}' saved securely to Vantuka Vault.")

    st.write("---")
    st.write("### Patient Database")
    p_name = st.text_input("Patient Name")
    p_age = st.number_input("Age", min_value=0)
    if st.button("Save Patient Profile"):
        st.info(f"Patient {p_name} added to global database.")

# --- 6. CONTACT & INQUIRIES ---
elif menu == "Contact & Inquiries":
    st.header("📩 Contact Us")

    with st.form("Inquiry Form"):
        name = st.text_input("Full Name")
        user_email = st.text_input("Email Address")
        msg = st.text_area("How can we help?")
        submit = st.form_submit_button("Send Inquiry")

        if submit:
            # Attempt to send the auto-reply email
            if send_auto_reply(user_email, name, SMTP_SENDER_EMAIL, SMTP_APP_PASSWORD):
                st.success("Inquiry Sent! An automatic email response has been triggered and sent to your email.")
            else:
                st.error("Failed to send auto-reply. Please check SMTP settings (sender email/app password) and your internet connection.")

            # Note: In a real app, you would also store the inquiry message in a database.
            st.write(f"**Auto-Reply Sent to {user_email}:** Hello {name}, thank you for contacting Vantuka Health. Our team will assist you shortly.")

    st.write(f"**Direct Contact:** {CONTACT_NO} | {EMAIL}")

# --- FOOTER ---
st.sidebar.write("---")
st.sidebar.write("🔒 **Security & Privacy Enabled**")
if st.sidebar.button("Logout"):
    st.sidebar.warning("Logged Out.")

def send_auto_reply(recipient_email, recipient_name, sender_email, app_password):
    # This function uses the provided sender_email and app_password for authentication
    msg = MIMEText(f"Hello {recipient_name},\n\nThank you for contacting Vantuka Health. "
                   "Our team has received your inquiry and will respond shortly.\n\n"
                   "Best regards,\nMatume Sebetola")
    msg['Subject'] = "Vantuka Health - Inquiry Received"
    msg['From'] = sender_email
    msg['To'] = recipient_email

    try:
        with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
            server.login(sender_email, app_password)
            server.sendmail(sender_email, recipient_email, msg.as_string())
        return True
    except Exception as e:
        # Log the error for debugging purposes
        print(f"Error sending email: {e}")
        return False
```
